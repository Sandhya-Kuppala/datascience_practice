# To create Hierarchical clustering for crimeData

```python
import pandas as pd
import matplotlib.pylab as plt
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch 
from sklearn.cluster import AgglomerativeClustering
```

```python
crime_data=pd.read_csv("crime_data.csv")
```

```python
crime_data.shape
crime_data.columns
```

```python
# Normalizing the data
def norm_func(i):
    x = (i-i.min())	/ (i.max()-i.min())
    return (x)
```

```python
crime_data.iloc[:, 1:]
```

```python
crime_norm = norm_func(crime_data.iloc[:, 1:])
```

```python
crime_norm
```

```python
# apply H-Clustering algorithm
dist = linkage(crime_norm, method = "complete", metric = "euclidean")
h_clust = AgglomerativeClustering(n_clusters = 3, linkage = 'complete', affinity = "euclidean").fit(crime_norm) 
h_clust.labels_
cluster_labels = pd.Series(h_clust.labels_)
```

```python
plt.figure(figsize=(15, 8));plt.title('Hierarchical Clustering Dendrogram');plt.xlabel('Index');plt.ylabel('Distance')
sch.dendrogram(dist,leaf_rotation = 0,leaf_font_size = 10)
plt.show()
```
